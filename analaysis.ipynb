```python
# Load dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/cleaned_customer_churn.csv")
df
```

```python
# descriptive analysis

df.groupby("churn")[["monthlycharges","tenure"]].mean()
```

```python
# Hypothesis Testing

# Monthly Charges vs Churn (t-test)

from scipy.stats import ttest_ind

churn_yes = df[df["churn"]=="Yes"]["monthlycharges"]
churn_no = df[df["churn"]=="No"]["monthlycharges"]

t_stat, p_value = ttest_ind(churn_yes, churn_no, equal_var=False)
p_value
```

```python
# Tenure vs Churn (t-test)

tenure_yes = df[df["churn"] == "Yes"]["tenure"]
tenure_no = df[df["churn"] == "No"]["tenure"]

t_stat, p_value = ttest_ind(tenure_yes, tenure_no)
p_value
```

```python
# Contract Type vs Churn (Chi-Square)

from scipy.stats import chi2_contingency

contingency = pd.crosstab(df["contract"], df["churn"])
chi2, p, _, _ = chi2_contingency(contingency)
p
```

```python
# Payment Method vs Churn

contingency = pd.crosstab(df["paymentmethod"], df["churn"])
chi2, p, dof, expected = chi2_contingency(contingency)
p
```
